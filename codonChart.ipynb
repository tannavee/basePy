```python
import pandas as pd
```

```python
# to keep track, wrote them row-wise, going along each row
codonR1 = ["UUU","UUC","UUA","UUG","UCU","UCC","UCA","UCG","UAU","UAC","UAA","UAG","UGU","UGC","UGA","UGG"]
AminoR1 = ["Phe","Phe","Leu","Leu","Ser","Ser","Ser","Ser","Tyr","Tyr","STOP","STOP","Cys","Cys","STOP","Trp"]

data1 = pd.DataFrame(codonR1, AminoR1).reset_index()
data1 = data1.rename(columns = {"index":"Amino Acid", 0:"Codon"})
data1.head()
```

```python
codonR2 = ["CUU","CUC","CUA","CUG","CCU","CCC","CCA","CCG","CAU","CAC","CAA","CAG","CGU","CGC","CGA","CGG"]
AminoR2 = ["Leu","Leu","Leu","Leu","Pro","Pro","Pro","Pro","His","His","Gln","Gln","Arg","Arg","Arg","Arg"]

data2 = pd.DataFrame(codonR2, AminoR2).reset_index()
data2  = data2.rename(columns = {"index":"Amino Acid", 0:"Codon"})

data2.head()
```

```python
codonR3 = ["AUU","AUC","AUA","AUG","ACU","ACC","ACA","ACG","AAU","AAC","AAA","AAG","AGU","AGC","AGA","AGG"]
AminoR3 = ["Ile","Ile","Ile","Met","Thr","Thr","Thr","Thr","Asn","Asn","Lys","Lys","Ser","Ser","Arg","Arg"]

data3 = pd.DataFrame(codonR3,AminoR3).reset_index()
data3 = data3.rename(columns={"index":"Amino Acid", 0:"Codon"})
data3.head()
```

```python
codonR4 = ["GUU","GUC","GUA","GUG","GCU","GCC","GCA","GCG","GAU","GAC","GAA","GAG","GGU","GGC","GGA","GGG"]
AminoR4 = ["Val","Val","Val","Val","Ala","Ala","Ala","Ala","Asp","Asp","Glu","Glu","Gly","Gly","Gly","Gly"]


data4 = pd.DataFrame(codonR4, AminoR4).reset_index()
data4 = data4.rename(columns={"index":"Amino Acid", 0:"Codon"})
data4.head()
```

```python
# put data frames together 
```

```python

codonChart = pd.concat([data1,data2,data3,data4], sort = False, axis = 0)
codonChart
```
